```python
rolls=random.choices([1,2,3,4,5,6],k=10)
mean=sum(rolls)/len(rolls)
mean
```

```python
my_list = []
for i in range(0,10):
    rolls=random.choices([1,2,3,4,5,6],k=10)
    mean=sum(rolls)/len(rolls)
    my_list.append(mean)
sns.histplot(my_list)
```
